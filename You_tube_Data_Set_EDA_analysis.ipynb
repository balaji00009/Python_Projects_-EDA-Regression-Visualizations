# 7. Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb' and so on.....


```python
# To assign the values of month as names as per requirement, the first step is to create the dict of month as values and numbers as keys
month_names = {1:"Jan",2:"Feb",3:"Mar",4:"Apr",5:"May",6:"Jun",7:"Jul",8:"Aug",9:"Sep",10:"Oct",11:"Nov",12:"Dec"}
```

```python
# By using the map function we can assign the month names as per requirement
df["published_month"] = df["published_month"].map(month_names)
```

```python
# checking the df, and the month assigned to the data frame successfully:-
df
```

# 14. Display all the information about the videos that were published in January, and mention the count of videos that were published in January

```python
# using filter function to show case the show the values in data frame:-
df[df["published_month"]=="Jan"]
#This clearly shows that it has about 2099 videos in the shape of rows
```

```python
# using for loop to get the count of videos that had been posted in the month of january:-
#initalizing the count to be zero:
count = 0
# using for loop to iterate over the published month:-
for i in df["published_month"]:
#checking if that iterator i which is equal to month Jan:-
    if i=="Jan":
# count will increased if value matches the given criteria:-
        count +=1
print(f"The count of videos which were published in January were about {count} videos")
```
